```python
dim1 = []
dim2 = []

file_path = os.path.join(test_path, 'uninfected')
for image_filename in os.listdir(file_path):
    img = imread(os.path.join(file_path, image_filename))
    d1, d2, colors = img.shape
    dim1.append(d1)
    dim2.append(d2)
```

```python
sns.jointplot(dim1, dim2)
```

```python
np.mean(dim1)
```
